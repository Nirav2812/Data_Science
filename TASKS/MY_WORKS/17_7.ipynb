```python
num = 13

for i in range(1, 11):
    expression = ""
    
    for j in range(i):
        expression += "13"
        if j != i - 1:
            expression += "+"
    
    print(f"{num}*{i}=({expression})={num*i}")
```

```python
num = int(input("enter the number = "))
fact = 1
i=1
while i<num+1:
    fact = fact*i
    i +=1
print(f"factorial of your number is {fact}")
```
