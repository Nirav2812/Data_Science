```python
def count_even(numbers):
    count=0
    for i in numbers:
        if i % 2 == 0 :
            count+=1
    return(count)
num = [1,2,4,3,5,6,7,8,9]

evens=count_even(num)

print(f"total number of even numbers is {evens}")
```

```python
def calculate_discount(price, rate=10) :
    disc=(price*rate)/100
    final_price=price-disc

    return final_price

print(calculate_discount(1000))
```

```python
def sales_summary(sales):
    ts=sum(sales)
    avgs=ts/len(sales)
    highest=0
    for i in sales:
        if i>highest:
            highest=i
    return(ts,avgs,highest)

sales=[200,2500,1450,1700,4200,3000]

total,avs,high=sales_summary(sales)

print(f"the total value sales is {total},total avg is {avs}, highest sale is {high}")

```

```python
def calculate_bonus(salary):
    if salary > 50000:
        return salary * 0.10
    else:
        return salary * 0.05

print(calculate_bonus(60000))
```
