```python
playlist1 = {"arijit", "sunilthakar","adityagadhvi","justin beber","taylor swift","sabrina carpenter"}
playlist2 = {"arijit","gadhvi","shreya ghosal","neha kakkar","honey sing","the weeknd",}

def get_unique_artists(spotify_playlist1, spotify_playlist2):
    all_artist=spotify_playlist1.union(spotify_playlist2)
    return(all_artist)

final_Set=get_unique_artists(playlist1,playlist2)

print(final_Set)
```
