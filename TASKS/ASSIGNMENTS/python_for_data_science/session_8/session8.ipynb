```python
steps = [6600, 5300, 10015, 9000, 12000, 12500, 10000]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

i = 0

while i < len(steps):
    if steps[i] > 10000:
        print("First day you crossed 10,000 steps:", days[i])
        break
    i += 1
```
