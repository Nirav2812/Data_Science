```python
def format_price(price,currency="INR"):
    price=str(price)
    # if currency=="INR":
    #     print(f"your bank balance is ₹ {price}")
    # else:
    #     print(f"your bank balance is $ {price}")
    if currency=="INR":
        st="₹"+ price
        return st
    else:
        st="$"+ price
        return st
```

```python
ans=format_price(1000)
print(f"your bank balance is $ {ans}")
```
